Mettre toutes les filiales dans un seul dataframe Data_Filiales
https://www.techbeamers.com/python-glob/

```python
import glob
```

```python
all_filiales = glob.glob('Remontees Filiales/*.csv')
vide = []
for i in all_filiales:
    df = pd.read_csv(i, index_col=None, header=0)
    vide.append(df)

Data_Filiales = pd.concat(vide, axis=0, ignore_index=True)
```

```python
Data_Filiales 
```
